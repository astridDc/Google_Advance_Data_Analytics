**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose to have Borough names as column headers
properties_transposed = properties.transpose()

# Reset the index to make the dates as a column and reset the column names
properties_transposed = properties_transposed.reset_index()

# Assign the values of the first row as column headings
properties_transposed.columns = properties_transposed.iloc[0]

# Drop the first row since it's now the column headers
properties_transposed = properties_transposed.drop(0)

# Check the first few rows 
properties_transposed.head()
```

```python
properties_transposed.columns
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Convert the index to a string and split on whitespace to extract only the date part
properties_transposed.columns = properties_transposed.columns.astype(str).str.split(' ').str[0]

properties_transposed.head()
```

```python
# Rename the 'NaT' column to 'ID'
properties_transposed = properties_transposed.rename(columns={'NaT': 'ID'})

properties_transposed.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt the DataFrame to create a tidy format
properties_melted = pd.melt(properties_transposed, id_vars=['Borough', 'ID'], var_name='Month', value_name='Average Price')

# Convert the 'Month' column to datetime
properties_melted['Month'] = pd.to_datetime(properties_melted['Month'])

# Drop rows with null values in the 'Average Price' column
properties_melted = properties_melted.dropna(subset=['Average Price'])

# Convert the 'Average Price' column to numeric
properties_melted['Average Price'] = pd.to_numeric(properties_melted['Average Price'])

# Display the first few rows 
print(properties_melted.head())
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(df_filter):
    # Get the average price for the years 1998 and 2018
    avg_price_1998 = df_filter[df_filter['Year'] == 1998]['Average Price'].mean()
    avg_price_2018 = df_filter[df_filter['Year'] == 2018]['Average Price'].mean()
    
    # Calculate the ratio of the average price for 1998 to 2018
    price_ratio = avg_price_1998 / avg_price_2018
    
    return price_ratio
```

```python
# Create an empty dictionary to store the price ratios for each borough
price_ratios = {}

# Iterate through each unique London borough
for borough in properties_filtered['Borough'].unique():
    # Get the filter for the current borough
    borough_filter = properties_filtered[properties_filtered['Borough'] == borough]
    
    # Calculate the price ratio for the current borough and store it in the dictionary
    price_ratios[borough] = create_price_ratio(borough_filter)

# Convert the dictionary to a DataFrame for better visualization
price_ratios_df = pd.DataFrame(price_ratios.items(), columns=['Borough', 'Price Ratio'])

# Sort the DataFrame by the price ratio in ascending order
price_ratios_df.sort_values(by='Price Ratio', ascending=True, inplace=True)

# Display the DataFrame with the price ratios for each borough
print(price_ratios_df)
```
